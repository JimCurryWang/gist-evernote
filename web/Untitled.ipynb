```python
images = [
    ('part_0.png', (0, 0)),
    ('part_1.png', (0, 1438)),
    ('part_2.png', (0, 1438 * 2))
]
```

```python
for image in images:
    file_name = image[0]
    offset = image[1]
    
    screenshot = Image.open(file_name)
    actual_width, actual_height = screenshot.size
    print(actual_width, actual_height)
    stitched_image.paste(screenshot, offset)
```
